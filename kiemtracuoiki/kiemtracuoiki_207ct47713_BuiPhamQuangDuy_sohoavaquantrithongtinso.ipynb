Kiểm tra định kỳ: Vì HR quản lý dữ liệu, họ sẽ là bộ phận chính chạy báo cáo giá trị thiếu, giúp phát hiện và bổ sung dữ liệu của chính mình.

```python
# Câu 2: Điền giá trị thiếu trong Performance bằng trung bình của Department
# Lưu dữ liệu Performance trước khi điền để vẽ histogram
performance_before = df['Performance'].copy()

# Tính trung bình Performance theo Department
mean_performance_by_dept = df.groupby('Department')['Performance'].mean()
print("\nCâu 2: Trung bình Performance theo Department:")
print(mean_performance_by_dept)

# Điền giá trị thiếu
df['Performance'] = df.apply(
    lambda row: mean_performance_by_dept[row['Department']] if pd.isna(row['Performance']) else row['Performance'],
    axis=1
)

# Lưu dữ liệu Performance sau khi điền để vẽ histogram
performance_after_dept = df['Performance'].copy()

# Đề xuất phương pháp điền khác dựa trên đặc điểm của Location
mean_performance_by_location = performance_before.groupby(df['Location']).mean()
print("\nTrung bình Performance theo Location (dữ liệu ban đầu):")
print(mean_performance_by_location)

# Điều chỉnh trung bình dựa trên đặc điểm Location
adjusted_means = {
    'Site A': mean_performance_by_location['Site A'] * 1.05,  # Tăng 5% do hiệu suất cao
    'Site B': mean_performance_by_location['Site B'] * 0.95,  # Giảm 5% do hiệu suất thấp
    'Site C': mean_performance_by_location['Site C']          # Giữ nguyên
}
print("\nTrung bình Performance điều chỉnh theo Location:")
print(pd.Series(adjusted_means))

# Điền thủ công theo Location
df['Performance_Location'] = performance_before.copy()
for idx, row in df.iterrows():
    if pd.isna(row['Performance_Location']):
        location = row['Location']
        df.loc[idx, 'Performance_Location'] = adjusted_means[location]

# Lưu dữ liệu Performance sau khi điền theo Location
performance_after_location = df['Performance_Location'].copy()

# Vẽ histogram trước và sau
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(performance_before, bins=10, edgecolor='black', alpha=0.7)
plt.title('Performance Trước khi Điền')
plt.xlabel('Performance')
plt.ylabel('Tần suất')

plt.subplot(1, 3, 2)
plt.hist(performance_after_dept, bins=10, edgecolor='black', alpha=0.7)
plt.title('Performance Sau khi Điền (Department)')
plt.xlabel('Performance')
plt.ylabel('Tần suất')

plt.subplot(1, 3, 3)
plt.hist(performance_after_location, bins=10, edgecolor='black', alpha=0.7)
plt.title('Performance Sau khi Điền (Location)')
plt.xlabel('Performance')
plt.ylabel('Tần suất')

plt.tight_layout()
plt.show()

```

Giải thích tại sao chiến lược này phù hợp với đặc điểm nhân viên:

- Site B: Giữ giờ làm việc nhưng tăng đào tạo giúp cải thiện kỹ năng, từ đó tăng hiệu suất mà không cần tăng giờ làm việc.
- Site C: Giảm giờ làm việc là cần thiết để tránh mệt mỏi, cải thiện hiệu suất bằng cách đảm bảo nhân viên có thời gian nghỉ ngơi.

```python
# Câu 4: Tính correlation giữa Hours_Worked và Performance cho từng Location
# Tạo dictionary để lưu correlation
correlation_by_location = {}

# Lọc dữ liệu theo từng Location và tính correlation
locations = df['Location'].unique()
for location in locations:
    # Lọc dữ liệu cho Location hiện tại
    df_location = df[df['Location'] == location][['Hours_Worked', 'Performance']]
    # Bỏ các hàng có giá trị NaN
    df_location = df_location.dropna()
    # Tính correlation
    if len(df_location) > 1:  # Cần ít nhất 2 giá trị để tính correlation
        correlation = df_location['Hours_Worked'].corr(df_location['Performance'])
        correlation_by_location[location] = correlation
    else:
        correlation_by_location[location] = np.nan

print("Câu 4: Correlation giữa Hours_Worked và Performance theo Location:")
for location, corr in correlation_by_location.items():
    print(f"{location}: {corr:.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Vẽ boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x="Department", y="Performance", hue="Gender", data=df)
plt.title("Boxplot của Performance theo Department và Gender")
plt.xlabel("Department")
plt.ylabel("Performance")
plt.legend(title="Gender")
plt.show()

# Xác định outlier bằng IQR
Q1 = df["Performance"].quantile(0.25)
Q3 = df["Performance"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Lọc outliers
outliers = df[(df["Performance"] < lower_bound) | (df["Performance"] > upper_bound)]
print("Nhân viên có Performance là outlier:")
print(outliers)
```

Câu 11: (4 điểm) Điền giá trị thiếu trong Performance bằng hồi quy tuyến tính thủ công dựa trên
Hours_Worked và Training_Hours. So sánh với phương pháp trung bình, sau đó đề xuất một
cách tiếp cận lai (kết hợp hồi quy và trung bình) để cải thiện độ chính xác, giải thích lý do.
Giải: 
Đề xuất phương pháp kết hợp (lai) để cải thiện độ chính xác
Cách tiếp cận lai:

Nếu dữ liệu có tương quan mạnh giữa Performance và (Hours_Worked, Training_Hours), dùng hồi quy tuyến tính.

Nếu dữ liệu không có mối quan hệ mạnh, dùng trung bình của Department thay vì trung bình toàn bộ.
Lý do phương pháp lai hiệu quả hơn
✅ Kết hợp cả mô hình dự đoán và trung bình theo phòng ban giúp giữ lại mối quan hệ tự nhiên trong dữ liệu.
✅ Hạn chế sai số của hồi quy tuyến tính nếu dữ liệu không có tương quan mạnh.
✅ Cải thiện độ chính xác của dữ liệu điền vào, tránh việc dự đoán quá xa thực tế.


```python
import numpy as np
import pandas as pd

# Sao chép DataFrame để xử lý
df_reg = df.copy()

# Lọc những dòng KHÔNG bị thiếu để huấn luyện mô hình hồi quy
train_data = df_reg.dropna(subset=['Performance', 'Hours_Worked', 'Training_Hours'])

# Lấy các biến độc lập (X) và biến mục tiêu (y)
X = train_data[['Hours_Worked', 'Training_Hours']].values
y = train_data['Performance'].values
# Bước 2: Hồi quy tuyến tính thủ công (tính hệ số bằng công thức)

# Dùng công thức giải tích:
# β = (XᵗX)⁻¹ Xᵗy
# Thêm cột hệ số chặn (bias) vào X
X_bias = np.c_[np.ones(X.shape[0]), X]  # X shape: (n_samples, 3)

# Tính hệ số beta
beta = np.linalg.inv(X_bias.T @ X_bias) @ X_bias.T @ y

# In hệ số hồi quy
print(f"Hệ số hồi quy (beta): {beta}")
# Bước 3: Điền giá trị thiếu bằng mô hình hồi quy
# Xác định các dòng bị thiếu Performance nhưng đủ Hours_Worked và Training_Hours
missing_perf = df_reg[
    df_reg['Performance'].isna() & 
    df_reg['Hours_Worked'].notna() & 
    df_reg['Training_Hours'].notna()
]

# Áp dụng mô hình để dự đoán Performance
for idx, row in missing_perf.iterrows():
    x_input = np.array([1, row['Hours_Worked'], row['Training_Hours']])  # thêm 1 để nhân với intercept
    prediction = x_input @ beta
    df_reg.at[idx, 'Performance'] = prediction  # Gán giá trị dự đoán vào DataFrame
# Bước 4: So sánh với phương pháp điền trung bình
# Tính trung bình Performance toàn bộ (bỏ NaN)
mean_perf = df['Performance'].mean()
print(f"Trung bình Performance toàn bộ: {mean_perf:.2f}")
```

Công bằng trong cơ hội phát triển: Đánh giá hiệu suất dựa trên các yếu tố khách quan và tạo cơ hội phát triển cho tất cả nhân viên, không phân biệt mức độ hiện tại của hiệu suất, giúp tạo ra môi trường công bằng và công nhận nỗ lực của tất cả nhân viên.

Thông qua quy trình này, công ty sẽ có thể nâng cao sự công bằng trong việc đánh giá và phát triển nhân viên, đồng thời giảm thiểu thiên vị và tạo ra một môi trường làm việc công bằng và thúc đẩy sự phát triển cá nhân.


```python
import pandas as pd
import numpy as np

# Hàm xác định nhân viên có Performance ngoài 2 độ lệch chuẩn
def identify_outliers(df):
    # Tính toán trung bình và độ lệch chuẩn của Performance
    mean_performance = df['Performance'].mean()
    std_performance = df['Performance'].std()
    
    # Xác định các nhân viên có Performance ngoài 2 độ lệch chuẩn
    outliers = df[(df['Performance'] > mean_performance + 2 * std_performance) | 
                  (df['Performance'] < mean_performance - 2 * std_performance)]
    
    return outliers

# Gọi hàm để tìm nhân viên có Performance ngoài 2 độ lệch chuẩn
outlier_employees = identify_outliers(df)

# Hiển thị kết quả
print("Nhân viên có Performance ngoài 2 độ lệch chuẩn:")
print(outlier_employees)
```

Câu 16: (10 điểm) Tải dữ liệu MNIST từ PyTorch (torchvision.datasets.MNIST). Chuẩn hóa pixel về
[0, 1], flatten thành vector 784 chiều. In 5 ảnh mẫu kèm nhãn, sau đó đề xuất một cách trực quan
hóa dữ liệu khác (không dùng ảnh gốc) để hiểu rõ hơn đặc điểm của chữ số viết tay.
Giải:
Đề xuất cách trực quan hóa khác:
PCA: Giảm chiều và trực quan hóa dữ liệu giúp chúng ta hiểu rõ hơn về sự phân bố của các chữ số trong không gian, đặc biệt là khi chúng ta không muốn nhìn vào ảnh gốc. PCA có thể giúp tìm ra các cấu trúc nội tại trong dữ liệu và phân nhóm các chữ số tương tự nhau.

```python
import torch
from torch import nn, optim
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np

# Bước 1: Tải dữ liệu MNIST và chuẩn hóa pixel về [0, 1]
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

trainset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
testset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)
testloader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle=False)

# Bước 2: In ra 5 ảnh mẫu kèm nhãn
# Lấy 5 ảnh đầu tiên trong tập huấn luyện
images, labels = next(iter(trainloader))

# Hàm vẽ 5 ảnh mẫu kèm nhãn
def imshow(images, labels):
    plt.figure(figsize=(10, 2))
    for i in range(5):
        plt.subplot(1, 5, i+1)
        plt.imshow(images[i].squeeze(), cmap='gray')
        plt.title(f'Label: {labels[i].item()}')
        plt.axis('off')
    plt.show()

# Vẽ 5 ảnh mẫu
imshow(images, labels)

# Bước 3: Đề xuất cách trực quan hóa dữ liệu khác
# Để hiểu rõ hơn đặc điểm của chữ số viết tay mà không dùng ảnh gốc, một cách tốt là sử dụng **principal component analysis (PCA)** để giảm chiều dữ liệu
# Chúng ta sẽ lấy một vài mẫu ảnh, flatten chúng thành vector 784 chiều và áp dụng PCA để giảm xuống 2 chiều để trực quan hóa.

# Flatten ảnh thành vector 784 chiều
images_flattened = images.view(images.size(0), -1)

# Dùng PCA để giảm chiều
from sklearn.decomposition import PCA

# Chỉ lấy một phần nhỏ dữ liệu để làm PCA nhanh chóng
X = images_flattened[:100].numpy()

# Áp dụng PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Vẽ 2D scatter plot của các chữ số
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels[:100].numpy(), cmap='tab10')
plt.colorbar(label='Digit Label')
plt.title('PCA - Projection of MNIST Data into 2D')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

Tùy vào đặc điểm của MNIST và tài nguyên máy tính, chiến lược chọn batch size có thể thay đổi để tối ưu hóa hiệu suất huấn luyện. Dưới đây là một số chiến lược và phân tích về ưu nhược điểm của chúng:

1. Batch size lớn (ví dụ: 128 hoặc 256)
Ưu điểm:

Việc sử dụng batch size lớn giúp giảm số lượng bước huấn luyện trong mỗi epoch, vì mỗi epoch sẽ cần ít bước hơn. Điều này có thể tiết kiệm thời gian huấn luyện, đặc biệt là khi xử lý với bộ dữ liệu lớn.

Các gradient được tính toán từ một lượng mẫu lớn sẽ ít bị nhiễu loạn và có thể dẫn đến sự hội tụ nhanh hơn.

Nhược điểm:

Sử dụng batch size lớn yêu cầu bộ nhớ GPU lớn, điều này có thể gây vấn đề nếu bộ nhớ không đủ.

Việc sử dụng batch size quá lớn có thể dẫn đến việc mô hình hội tụ chậm và không tốt cho việc tối ưu hóa (vì các bước gradient có thể ít đa dạng, dẫn đến việc bỏ qua các điểm tối ưu nhỏ).

2. Batch size nhỏ (ví dụ: 32 hoặc 16)
Ưu điểm:

Batch size nhỏ giúp mô hình có thể học được nhiều đặc điểm của dữ liệu hơn, vì mỗi bước huấn luyện sẽ có sự biến động lớn hơn. Điều này có thể giúp mô hình tránh bị mắc kẹt trong các điểm tối ưu local.

Tiêu thụ ít bộ nhớ GPU hơn, vì mỗi batch yêu cầu ít dữ liệu hơn.

Nhược điểm:

Số lượng bước huấn luyện trong mỗi epoch sẽ nhiều hơn, điều này có thể làm tăng thời gian huấn luyện.

Các gradient tính toán từ batch nhỏ có thể bị nhiễu loạn nhiều hơn, dẫn đến việc mô hình có thể không hội tụ nhanh như khi sử dụng batch size lớn.

3. Batch size tự động điều chỉnh (dynamic batch size)
Một chiến lược có thể áp dụng là batch size tự động điều chỉnh. Theo đó, batch size sẽ được thay đổi tùy vào mức độ tài nguyên còn lại trong hệ thống. Ví dụ:

Khi bộ nhớ GPU còn nhiều, tăng batch size.

Khi bộ nhớ GPU gần đầy, giảm batch size.

Ưu điểm:

Tối ưu hóa tài nguyên máy tính, giúp giảm thời gian huấn luyện trong khi vẫn đảm bảo huấn luyện hiệu quả.

Nhược điểm:

Phức tạp hơn trong việc quản lý và cấu hình, vì cần phải thay đổi batch size trong quá trình huấn luyện.

4. Mini-batch gradient descent (với batch size 64)
Ưu điểm:

Batch size 64 là một lựa chọn khá phổ biến, có sự cân bằng giữa tốc độ và tính ổn định. Nó vừa đủ nhỏ để giảm thiểu việc tiêu thụ bộ nhớ, vừa đủ lớn để đảm bảo hiệu quả huấn luyện tốt.

Dễ dàng tối ưu hóa và kiểm soát, vì đây là lựa chọn mặc định trong nhiều trường hợp.

Nhược điểm:

Không thể tối ưu hóa tài nguyên máy tính như các chiến lược dynamic batch size.




```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split

# Bước 1: Tải dữ liệu MNIST và chuẩn hóa pixel về [0, 1]
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

trainset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)

# Chia tập train thành 80% train và 20% test
train_size = int(0.8 * len(trainset))
test_size = len(trainset) - train_size
train_subset, val_subset = random_split(trainset, [train_size, test_size])

# Bước 2: Tạo DataLoader với batch size 64
batch_size = 64

trainloader = DataLoader(train_subset, batch_size=batch_size, shuffle=True)
valloader = DataLoader(val_subset, batch_size=batch_size, shuffle=False)

# Kiểm tra thông tin
print(f"Total training samples: {len(train_subset)}")
print(f"Total validation samples: {len(val_subset)}")
print(f"Batch size: {batch_size}")
```

Kiến trúc đề xuất:
	•	Tăng số lượng nơ-ron ẩn: 784 → 256 → 128 → 64 → 10
	•	Thêm Dropout sau mỗi tầng ReLU để giảm overfitting
    # Ví dụ cấu trúc mới trong PyTorch:
        nn.Sequential(
            nn.Linear(784, 256),
            nn.ReLU(),
            nn.Dropout(0.3),        # Dropout để tránh overfitting
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 10)
        )
Giải thích lý do:
	•	Việc tăng số nơ-ron giúp mô hình biểu diễn tốt hơn các đặc trưng phức tạp trong dữ liệu.
	•	Dropout giúp giảm hiện tượng học thuộc (overfitting), đặc biệt trên dữ liệu đơn giản như MNIST.
	•	Kiến trúc sâu hơn giúp mô hình học được quan hệ phi tuyến phức tạp hơn, từ đó cải thiện độ chính xác.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np

# Bước 1: Chuẩn bị dữ liệu MNIST
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

trainset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
testset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

trainloader = DataLoader(trainset, batch_size=64, shuffle=True)
testloader = DataLoader(testset, batch_size=64, shuffle=False)

# Bước 2: Xây dựng kiến trúc ANN
class ANN(nn.Module):
    def __init__(self):
        super(ANN, self).__init__()
        self.fc1 = nn.Linear(28*28, 128)  # 784 -> 128
        self.fc2 = nn.Linear(128, 64)  # 128 -> 64
        self.fc3 = nn.Linear(64, 10)  # 64 -> 10 (output layer)
        self.relu = nn.ReLU()
        self.softmax = nn.Softmax(dim=1)

    def forward(self, x):
        x = x.view(-1, 28*28)  # Flatten input (batch_size, 28*28)
        x = self.relu(self.fc1(x))  # First hidden layer
        x = self.relu(self.fc2(x))  # Second hidden layer
        x = self.fc3(x)  # Output layer
        return x

# Bước 3: Khởi tạo mô hình, loss function và optimizer
model = ANN()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Bước 4: Early stopping
patience = 5
best_loss = np.inf
patience_counter = 0

train_losses = []
test_losses = []

# Bước 5: Huấn luyện mô hình
num_epochs = 20
for epoch in range(num_epochs):
    model.train()
    running_train_loss = 0.0
    for images, labels in trainloader:
        optimizer.zero_grad()  # Clear previous gradients
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_train_loss += loss.item()

    avg_train_loss = running_train_loss / len(trainloader)
    train_losses.append(avg_train_loss)

    # Kiểm tra mô hình trên tập test
    model.eval()
    running_test_loss = 0.0
    with torch.no_grad():
        for images, labels in testloader:
            outputs = model(images)
            loss = criterion(outputs, labels)
            running_test_loss += loss.item()

    avg_test_loss = running_test_loss / len(testloader)
    test_losses.append(avg_test_loss)

    # Early stopping: kiểm tra nếu loss giảm
    if avg_test_loss < best_loss:
        best_loss = avg_test_loss
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f'Early stopping at epoch {epoch+1}')
            break

    # In thông tin huấn luyện mỗi epoch
    print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {avg_train_loss:.4f}, Test Loss: {avg_test_loss:.4f}')

# Bước 6: Vẽ biểu đồ loss
plt.plot(train_losses, label='Train Loss')
plt.plot(test_losses, label='Test Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Test Loss over Epochs')
plt.legend()
plt.show()
```

Scaling: Thay đổi kích thước của ảnh mà không thay đổi tỷ lệ.

Elastic distortions: Làm biến dạng ảnh bằng các mô hình co giãn để mô phỏng các kiểu viết tay khác nhau.
